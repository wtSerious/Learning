### 卷积的三个重要根据
- Sparse interactions
- Parameters sharing
- equivariant representations

#### Sparse interationsinput
传统的神经网络中，每一个 output unit 都会对应所有的input unit，那么如果 input unit 的数量很大，比如一张 100*100 的图片，它的输入层的单元数就为 10000，那么 hidden unit 和 output unit 的数量也会大，网络层数多的话对十分难以计算。而在卷积中利用通过 Sparse interactions 可以减少参数的数量级，提高计算效率并且带来好的效果。

在图 9.3 中我们可以发现，经过卷积以后，一个 output unit 不再是由所有的 input uinit决定了，一个 output uinit 由三个 input unit 影响，一个 input uinit 只影响 三个 output uinit。这就叫作 Sparse interactions。 Sparse interactions 能带来哪些好处呢。最直观的就是它减少了参数数量，提高了计算效率。但这个处理会不会带来什么问题呢，原本采用的是全连接方式，现在采用 Sparse interactions 后 output uinit 与 input init 之间的联系不再那么紧密，会不会再来一些不好的影响呢？在图 9.4 中我们可以看到，在 CNN 更深层中的 unit 虽然他们的直接连接使很稀疏的，但是深层的单元能间接的连接在全部或者大部分 input uinit。

#### Parameter sharing
这里的 Parameter sharing 其实指的是 weights sharing。是权重的共享，在传统的神经网络中，假设一个 $n*m$ 的 $W$ ,则其中一个 $W_{ij}$ 对应输入 $X_{ij}$， 每一个 $W_{ij}$ 在矩阵乘法运算中只被用到了一次。而在卷积中用 filter 当作权重，并且每一个 $filter_{ij}$ 被用于输入的每一个位置，权重参数被重复使用，每一个位置都用了相同的参数，这就叫作 Parmeter sharing。
